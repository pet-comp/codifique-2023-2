##Casting

É uma transformação forçada em valores numéricos para modificar seu tipo de dado.

Exemplo: Vamos transformar o resto de uma equação de float em int

```python
num = float(input("digite seu dinheiro "))

print("seu número arredondado é", int(num + 1))
```

#Operadores

Após aprendermos entradas, saidas e variáveis, depois de recebermos entradas numéricas diferentes vamos aprender oque fazer com essas entradas utilizando operadores, eles podem ser divididos em operadores aritmédicos e operadores de atribuição.

##Operadores Aritméticos
- " + " - Adição
- " - " - Subtração
- " * " - Multiplicação
- " / " - Divisão
- " // " - Divisão Inteira
- " % " - Resto da Divisão
- " ** " - Exponenciação

```python
idade_filho1 = 10
idade_filho2 = 11

idade_soma = idade_filho1 + idade_filho2

print("A soma das idades é", idade_soma)
```

##Operadores Atribuição
- " = " - igualdade
- " += " - soma com a variável
- " -= " - subtrai com a variável
- " *= " - multiplica com avariável
- " /= " - divide com a variável
- " %= " - resto da divisão com a variável

```python
idade_filho1 = 10
idade_filho2 = 30

print("A sua idade é ", idade_filho1)

idade_filho1 += idade_filho2

print("A sua idade agora é ", idade_filho1)
```
