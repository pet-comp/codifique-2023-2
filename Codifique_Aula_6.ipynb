### Cria√ß√£o de uma lista

**Criar uma lista** √© bem simples, para isso basta colocar os elementos que voc√™ deseja armazenar entre colchetes (**[ ]**)


````python
# aqui estamos criando uma lista vazia
lista_vazia = []

# aqui estamos criando uma lista composta por 4 elementos
lista = ['p√£o', 'abacaxi', 'uva', 'queijo']

# vamos ver se as listas foram criadas corretamente
print(lista_vazia, lista)
````

### Acesso √† elementos (indexa√ß√£o)

Ok, voc√™ j√° viu que listas s√£o um *conjunto de elementos* em uma vari√°vel s√≥, ent√£o agora vamos te mostrar como acessar um (ou mais) elementos espec√≠ficos de uma lista.
<br>
<br>

Antes disso vamos falar de como as posi√ß√µes (que voc√™ tamb√©m pode ver por a√≠ com o nome chic de *√≠ndice*) s√£o organizadas. Bom, como comentado anteriormente, as listas s√£o conjuntos *ordenados* de elementos, sendo que o primeiro elemento vai sempre estar na posi√ß√£o 0, o segundo da posi√ß√£o 1, o terceiro na posi√ß√£o 2 e assim segue. Logo, se voc√™ deseja acessar o primeiro elemento deve utilizar a posi√ß√£o 0, o quarto a posi√ß√£o 3 etc.

<center><img src='https://i.ibb.co/3MB1wwt/pos-index.jpg' width=600></center>

Fazer isso com Python √© bem simples, basta fazer `nome_da_lista[posi√ß√£o]`. Vamos ver um exemplo pr√°tico:

````python
# cria√ß√£o da lista
lista = ['p√£o', 'abacaxi', 'uva', 'queijo']

# aqui estamos acessando terceiro elemento da nossa lista passando o seu √≠ndice (2)
lista[2]
````

````python
# aqui vamos acessar o primeiro elemento da lista (√≠ndice 0)
lista[0]
````

Uma coisa que pode ser feita tamb√©m √© acessar mais de um elemento de uma vez por meio de algo que chamamos de *slicing*. A ideia √© bem parecida, sendo que podemos usar o seguinte comando `noem_da_lista[posi√ß√£o incial : posi√ß√£o final]` para pegar todos os elementos da posi√ß√£o incial at√© a posi√ß√£o final (sendo o elemento dessa √∫ltima posi√ß√£o n√£o incluso). Ou seja, se escrevemos `lista[1:3]` os elementos que s√£o retornados s√£o 'abacaxi' e 'uva'.

<center><img src='https://i.ibb.co/4YddBj4/slicing.jpg' width=600></center>

````python
# exemplos pr√°ticos
print('Exemplo 1:')
print(lista[1:3])

print('Exemplo 2:')
print(lista[0:2])

print('Exemplo 3:')
print(lista[1:4])
````

### Inser√ß√£o e remo√ß√£o de elementos

Ok, vamos voltar ao nosso exemplo da lista de compras agora. Imagine que voc√™ quer adicionar um elemento a sua lista, uma maneira de fazer isso √© atrav√©s do comando `nome_da_lista.append()`.
<br>
<br>


> O comando `.append` √© utilizado para adicionar elementos no final da lista



````python
# lista de compras
lista = ['p√£o', 'abacaxi', 'uva', 'queijo']

# vamos adicionar o elemento 'suco de laranja' no final da lista
lista.append('suco de laranja')

# verificamos que o elemento foi adicionado
print(lista)
````

Uma coisa que voc√™ pode fazer tamb√©m √© adicionar um novo elemento em uma posi√ß√£o espec√≠fica da lista em vez de simplesmente acrescentar no final, sendo que isso pode ser feito por meio do comando `nome_da_lista.insert(posi√ß√£o, elemento)`.
<br>
<br>


> O comando `.insert(posi√ß√£o, elemento)` √© utilizado para adicionar elementos em uma posi√ß√£o espec√≠fica


````python
print(f"Lista antes da inser√ß√£o: {lista}")

# vamos adicionar o elemento 'bolacha' no √≠ndice 2
lista.insert(2, 'bolacha')

print(f"Lista depois da inser√ß√£o: {lista}")
````

Para remover o elemento da √∫ltima posi√ß√£o da lista (ou de uma posi√ß√£o espec√≠fica) podemos utilizar o comando `.pop()`

````python
lista = ['p√£o', 'abacaxi', 'uva', 'queijo']

print(f"Lista antes da remo√ß√£o: {lista}")

# vamos remover o √∫ltimo elemento da lista
lista.pop()

print(f"Lista depois da remo√ß√£o: {lista}")
````

````python
lista = ['p√£o', 'abacaxi', 'uva', 'queijo']

print(f"Lista antes da remo√ß√£o: {lista}")

# vamos remover o √∫ltimo elemento da posi√ß√£o 2
lista.pop(2)

print(f"Lista depois da remo√ß√£o: {lista}")
````

Por √∫ltimo, para remover um elemento espec√≠fico, podemos utilizar o comando `.remove(elemento)`

````python
lista = ['p√£o', 'abacaxi', 'uva', 'queijo']

print(f"Lista antes da remo√ß√£o: {lista}")

# vamos remover o elemento 'p√£o'
lista.remove('p√£o')
# em seguida vamos remover o elemento 'uva'
lista.remove('uva')

print(f"Lista depois da remo√ß√£o: {lista}")
````

### Usando listas dentro de loops

Lembra dos nossos queridos loops, que percorrem por v√°rios elementos, um elemento de cada vez. Ent√£o, pensa comigo, listas nada mais s√£o que um conjunto de elementos n√£o √© mesmo? Ent√£o caso a gente queira acessar um elemento da lista de cada vez podemos pass√°-la para um for! Bora ver como isso funciona.

````python
lista = ['p√£o', 'abacaxi', 'bolacha', 'uva', 'queijo', 'suco de laranja']

# aqui n√≥s vamos pegar um elemento da lista em cada itera√ß√£o e imprim√≠-lo
for elemento in lista:
  print(elemento)

print("---------------------------")

# outra maneira de fazer isso √© atrav√©s dos √≠ndices (apesar da primeira maneira ser mais comum)
for indice in range(6):
  print(lista[indice])
````

## **Dicion√°rios**

Para entender o que s√£o dicion√°rios vamos pensar em outra analogia: imagine que voc√™ tem um conjunto de arm√°rios trancados, sendo que voc√™ guardou seus materiais escolares em um deles. Para conseguir acessar o que est√° dentro do arm√°rio n√£o adianta saber o n√∫mero do arm√°rio onde voc√™ guardou as suas coisas, voc√™ tamb√©m precisa ter a *chave* do arm√°rio que deseja abrir para conseguir acess√°-lo e pegar as suas coisas.

De maneira similar, dicion√°rios s√£o estruturas que relacionam chaves com valores, de maneira que para acessar um valor espec√≠fico √© necess√°rio utilizar a chave espec√≠fica associada √†quele valor.

Por enquanto isso pode n√£o estar t√£o claro, mas relaxa, vamos ver alguns exemplos aqui pra no final voc√™ estar entendendo criando dicion√°rios at√© de olho fechado!

### Cria√ß√£o de um dicion√°rio

Para criar uma lista vimos que s√£o utilizados os colchetes [ ], j√° os dicion√°rios s√£o representados pelas chaves (**{ }**). Os dicion√°rios s√£o formados por conjuntos de `chave : valor` (a chave e o valor associado a ela s√£o separados por ":"). Sendo que eles podem ser escritos como `dicionario = {chave_1 : valor_1, chave_2 : valor_2, chave_3 : valor_3, ...}`

Primeiramente, vamos criar um dicion√°rio que associa v√°rias pessoas (chaves) √†s suas respectivas idades (valores).

<br>
<br>
<table>
  <tr>
    <th>Pessoa</th>
    <th>Idade</th>
  </tr>
  <tr>
    <td>Laura</td>
    <td>16</td>
  </tr>
    <tr>
    <td>Matheus</td>
    <td>14</td>
  </tr>
    <tr>
    <td>Roberta</td>
    <td>21</td>
  </tr>
    <tr>
    <td>Lucas</td>
    <td>19</td>
  </tr>
    <tr>
    <td>Ana</td>
    <td>15</td>
  </tr>
</table>

````python
# cria√ß√£o de um dicion√°rio vazio
dicionario_vazio = {}

# cria√ß√£o de um dicion√°rio com 5 elementos que associa cada pessoa a sua idade
dicionario = {"Laura" : 16, "Matheus" : 14, "Roberta" : 21, "Lucas" : 19, "Ana" : 15}

# vamos verificar que os dicion√°rios foram criados corretamente
print(dicionario_vazio, dicionario)
````

### Acesso √† elementos

Para acessar elementos de um dicion√°rio n√≥s n√£o vamos utilizar os seus √≠ndices/posi√ß√µes como fizemos anteriormente com as listas, em vez disso n√≥s vamos acessar os valores por meio das suas chaves (assim como voc√™ acessa os items dentro de um arm√°rio utilizando a sua chave), ou seja, fazemos `dicionario[chave]`. Por exemplo, se queremos verificar qual √© a idade da Ana, podemos acessar esse valor escrevendo `dicionario["Ana"]`. Vamos ver alguns exemplos:

````python
dicionario = {"Laura" : 16, "Matheus" : 14, "Roberta" : 21, "Lucas" : 19, "Ana" : 15}

dicionario["Ana"]
````

````python
dicionario["Matheus"]
````

Outra coisa que podemos fazer √© obter uma lista de todas as chaves de um dicion√°rio, de todos os valores, ou de todos os itens. Isso pode ser feito, respectivamente, pelos comandos `dicionario.keys()`, `dicionario.values()` e `dicionario.items()`.

````python
dicionario.keys()
````

````python
dicionario.values()
````

````python
dicionario.items()
````

### Inser√ß√£o e remo√ß√£o de elementos

Para acrescentar elementos ao nosso dicion√°rio fazemos `dicionario[nova chave] = novo valor` e a dele√ß√£o de elementos tamb√©m √© bem simples, basta usar o comando `del(dicionario[chave a ser deletada])`. Agora bora pros exemplos!

````python
dicionario = {"Laura" : 16, "Matheus" : 14, "Roberta" : 21, "Lucas" : 19, "Ana" : 15}

print(f"Dicion√°rio antes da adi√ß√£o: {dicionario}")

# vamos acrescentar um novo elemento ao dicion√°rio, a Melissa que possui 16 anos
dicionario['Melissa'] = 16
# vamos acrescentar tamb√©m o Wesley de 17 anos
dicionario['Wesley'] = 17

print(f"Dicion√°rio depois da adi√ß√£o: {dicionario}")
````

````python
print(f"Dicion√°rio antes da dele√ß√£o: {dicionario}")

# vamos remover a Roberta
del(dicionario['Roberta'])
# vamos remover a Ana tamb√©m
del(dicionario['Ana'])

print(f"Dicion√°rio depois da dele√ß√£o: {dicionario}")
````

### Usando dicion√°rios dentro de loops

A ideia de usar dicion√°rios dentro de loops √© a mesma que de listas, ou seja, como temos v√°rios elementos dentro de um dicion√°rio n√≥s podemos iterar por eles, um por um.
<br>
<br>
Mas como em dicion√°rios n√≥s conseguimos acessar tanto apenas chaves (`.keys()`), quanto apenas valores (`.values()`) ou os itens inteiros (`.items()`) existem mais maneiras de construir os nossos loops (mas todas s√£o bem parecidas).

````python
dicionario = {'Laura': 16, 'Matheus': 14, 'Roberta': 21, 'Lucas': 19, 'Ana': 15, 'Melissa': 16, 'Wesley': 17}

# iterando pelos itens
for chave, valor in dicionario.items():
  print(f"{chave} possui {valor} anos")

print('--------------------------------')

# iterando apenas pelas chaves
for chave in dicionario.keys():
  print(chave)

print('--------------------------------')

# iterando pelos valores
for valor in dicionario.values():
  print(valor)

print('--------------------------------')

# tamb√©m podemos iterar pelos valores de uma maneira diferente (apesar da recomendada ser utilizando .values())
for chave in dicionario.keys():
  print(dicionario[chave])
````
